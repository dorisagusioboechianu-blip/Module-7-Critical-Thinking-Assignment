```python
"""START

Create following dictionaries room_nums,instructors,meeting_times

PROMPT user to enter a course id
CONVERT user input to uppercase and remove extra spaces

IF course exists in the room_nums dictionary:
   Display:
     Course Information:
     Course Number
     Room Number
     Instructor
     Meeting Time
Else:
   Display "Course not found"
END
"""

```

```python
room_numbers = {
    "CSC101": "3004",
    "CSC102": "4501",
    "CSC103":"6755",
    "NET110":"1244",
    "COM241":"1411"

}
instructors= {
    "CSC101": "Haynes",
    "CSC102": "Alvarado",
    "CSC103":"Rich",
    "NET110":"Burke",
    "COM241":"Lee"

}

meeting_times = {
    "CSC101": "8:00 am",
    "CSC102": "9:00 am",
    "CSC103":"10:00 am",
    "NET110":"11:00 am",
    "COM241":"1:00 pm"

}


course_id = input("Enter course id: ").strip().upper()

if course_id in room_numbers:
  print("\nCourse Information:")
  print("--------------------")
  print(f"Course Number: {course_id}")
  print(f"Room Number: {room_numbers[course_id]}")
  print(f"Instructor : {instructors[course_id]}")
  print(f"Meeting Times: {meeting_times[course_id]}")
else:
  print("Course not found")



```
